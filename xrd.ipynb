## 3 premiers pics du diffractogramme associé au matériau ZrFeTe pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
with MPRester("j1twApjMu9mGh9tITSd6") as m: 
    struct = m.get_structure_by_material_id("mp-961649")
    
xrdcalc = XRDCalculator(wavelength = 'CuKa', symprec = 0, debye_waller_factors = None)
xrdcalc.show_plot(struct)

xrdpattern = xrdcalc.get_pattern(structure = struct, scaled = True, two_theta_range = (0,45))

theta = xrdpattern.x #Tableau contenant les angles 2*theta
hkl = xrdpattern.hkls #Tabeau de tableaux contenant les différents indices hkl ainsi que leur multiplicité
intensity = (xrdpattern.intensity)[0:3] #intensité associée aux 3 premiers pics

print("\n Le premier pic sera observé avec un angle 2θ =",theta[0],"° et correspond aux plans",hkl[0][0]["hkl"]," et ",hkl[0][1]["hkl"])
print("L'intensité associée est de ", str(intensity[0]),"%")
print("\n Le second pic sera observé avec un angle 2θ =",theta[1],"° et correspond au plan",hkl[1][0]["hkl"])
print("L'intensité associée est de ", str(intensity[1]),"%")
print("\n Le troisième pic sera observé avec un angle 2θ =",theta[2],"° et correspond aux plans",hkl[2][0]["hkl"]," et ",hkl[2][1]["hkl"])
print("L'intensité associée est de ", str(intensity[2]),"%")
```
